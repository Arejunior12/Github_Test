```python
def sum_ave(a,b,c,d,e):
    f = a+b+c+d+e
    g = f/5
    return f, g
```

```python
sum_ave(2,4,6,8,5)
```

```python
def sum_ave(*list):  # *kwargs
    f = sum(list)
    g = f/len(list)
    return f, g
```

```python
sum_ave(2,3,5,5,76,4,3,2,1,3,4,5,6,7,7,7,6,4,3,2,2,3,5,6,6)
```

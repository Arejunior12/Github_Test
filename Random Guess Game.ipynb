### 2. Adding Score for Correct Guesses

```python
# Initialize score
score = 0

# Generate a random number between 1 and 10
random_number = random.randint(1, 10)

print("Guess the number (between 1 and 10):")

# Get user input
user_guess = int(input())

# Check if the guess is correct
if user_guess == random_number:
    print("Congratulations! You guessed it right.")
    score += 10
    #score = score + 10
else:
    print(f"Sorry, the correct number was {random_number}.")
    score -= 10

print(f"Your score: {score}")
```

### 3. Multiple Trials

```python
# Generate a random number between 1 and 10
random_number = random.randint(1, 10)

print("Guess the number (between 1 and 10). You have 3 attempts:")

attempts = 3
while attempts > 0:  # True
    user_guess = int(input(f"Attempt {4 - attempts}: "))
    if user_guess == random_number:
        print("Congratulations! You guessed it right.")
        break
    elif user_guess < random_number:
        print("Too low!")
    else:
        print("Too high!")
    attempts -= 1
    #attempts = attempts - 1

if attempts == 0:
    print(f"Out of attempts! The correct number was {random_number}.")
```

### 4. Hint System

```python
# Generate a random number between 1 and 10
random_number = random.randint(1, 20)

print("Guess the number (between 1 and 20). You have 5 attempts:")

attempts = 5
while attempts > 0:
    user_guess = int(input(f"Attempt {6 - attempts}: "))
    difference = abs(random_number - user_guess)

    if user_guess == random_number:
        print("Congratulations! You guessed it right.")
        break
    elif difference <= 2:
        print("Very close!")
    elif user_guess < random_number:
        print("Too low!")
    else:
        print("Too high!")
    attempts -= 1

if attempts == 0:
    print(f"Out of attempts! The correct number was {random_number}.")
```

### 5. Scoring with Dynamic Range and Replay Option

```python
def play_game():
    print("Choose difficulty: Easy (1-10), Medium (1-50), Hard (1-100)")
    difficulty = input("Enter 'easy', 'medium', or 'hard': ").lower() # Easy

    if difficulty == 'easy':
        max_number = 10
    elif difficulty == 'medium':
        max_number = 50
    elif difficulty == 'hard':
        max_number = 100
    else:
        print("Invalid choice, defaulting to Easy.")
        max_number = 10

    random_number = random.randint(1, max_number)
    print(f"Guess the number (between 1 and {max_number}). You have 7 attempts:")

    attempts = 7
    score = 0  # initial score

    while attempts > 0:
        user_guess = int(input(f"Attempt {8 - attempts}: "))
        if user_guess == random_number:
            print("Congratulations! You guessed it right.")
            score += 10 * attempts  # Reward more points for fewer attempts
            break
        elif user_guess < random_number:
            print("Too low!")
        else:
            print("Too high!")
        attempts -= 1

    if attempts == 0:
        print(f"Out of attempts! The correct number was {random_number}.")
    print(f"Your score: {score}")
    return score

# Main loop
total_score = 0
while True:
    total_score += play_game()
    replay = input("Do you want to play again? (yes/no): ").lower()
    if replay != 'yes':
        print(f"Thanks for playing! Your total score: {total_score}")
        break
```
